## Experiment 2 - transition cost

```python
import os

os.chdir("../codes")

from ZUI_MDP_solution import *

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
steps = 50
X = np.linspace(0,0.98, steps + 1)
Xs = (1/(1-X)) - 1

policies = []
policy_changes = []
old_policy = None
for i in range(steps + 1):
    value = Xs[i]
    gw = GridWorld.get_world('5x5', action_cost = value)
    Q = gw.policy_iteration()
    V = gw.Q2V(Q)
    policy = gw.Q2policy(Q)
    
    if not old_policy is None:
        if not np.array_equal(policy, old_policy):
            policy_changes.append(value)
            policies.append((policy, V, value))
    else:
        policies.append((policy, V, value))
    old_policy = policy

plt.plot(X,Xs)
plt.xlim([0, 1.0])
plt.savefig('../report/figures/e2_trans.pdf', dpi=500, transparent=True)
plt.show()
    
print(policy_changes)
```

```python
policy, V, value = policies[10]

gw = GridWorld.get_world('5x5', action_cost = value)
gw.plot()
plt.savefig('../report/figures/e2_board.pdf', dpi=500, transparent=True)
plt.show()
gw.plot(V = V,policy=policy)
plt.savefig('../report/figures/e2_policy_1.pdf', dpi=500, transparent=True)
plt.show()
print(value)
```
