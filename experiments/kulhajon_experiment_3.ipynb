## Experiment 3 - heatmap

```python
import os
os.chdir("../codes")
from ZUI_MDP_solution import *

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

gw = GridWorld.get_world('3x4')
gw.plot()
```

```python
steps = 10

X = np.linspace(0,1.0, steps + 1)
Y = np.linspace(0,3.0, steps + 1)

value_mask = [0, 3, 6, 8, 9, 10, 11]

Z = np.zeros((steps + 1, steps + 1, len(value_mask)), dtype = np.float32)
for i in range(steps + 1):
    for j in range(steps + 1):
        value = X[i]
        gw = GridWorld.get_world('3x4', action_proba = X[i], action_cost = Y[j])
        Q = gw.policy_iteration()
        V = gw.Q2V(Q)
        policy = gw.Q2policy(Q)

        Z[i,j,:] = V[value_mask]
```

```python
plt.contourf(X,Y,Z[:,:,5],15,cmap=plt.cm.rainbow)
plt.colorbar()
plt.savefig('../report/figures/e3_1.pdf', dpi=500, transparent=True)
plt.xlabel('Probability')
plt.ylabel('Transition cost')
plt.show()
```
